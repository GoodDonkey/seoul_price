### 올바른 시군구 코드가 있는 데이터를 불러와 보았다.

```python
df_si = pd.read_csv('../dataset/adm_code.csv')
df_si_seoul = df_si[df_si['시도명칭'] == '서울특별시']
df_si_seoul
```

```python
# 해당 구의 시군구번호가 몇번인지 확인해보자
df_si_seoul[df_si_seoul['시군구명칭']=='강남구']['시군구코드'].unique()[0]
```

### 그다음 for문을 돌려서 내 목표 df의 시군구 코드를 변경한다.
그 전에 구 이름들을 확인해본다

```python
df['geo_name'].unique()
```

```python
df_si_seoul['시군구명칭'].unique()
```

```python
# 타입을 바꿔준다. 안해줘도 될듯
df_si_seoul['시군구코드'] = df_si_seoul['시군구코드'].astype(str)
```

```python
df_si_seoul.info()
```

```python
# 동작구를 예로들어 확인해본다.
df_si_seoul[df_si_seoul['시군구명칭']=='동작구']['시군구코드']
```

```python
# loc으로 인덱싱 해야 바뀐다....는걸 깨달았다.
df.loc[(df['geo_name'] == '동작구'),'region_id'] = '11200'
df[df['geo_name']=='동작구']
```

- geo_name에 맞는 region_id를 변경하는 for문 코드

```python
for i in df_si_seoul['시군구명칭'].unique():
    df.loc[(df['geo_name'] == i),'region_id'] = \
    df_si_seoul.loc[(df_si_seoul['시군구명칭']== i),'시군구코드'].unique()[0]
```

```python
df[df['geo_name'] == '강남구']['region_id']
```

```python
df.to_csv('/Users/motive/Data_Study/202105_lab/seoul_price/dataset/seoul_id_alliance')
```
